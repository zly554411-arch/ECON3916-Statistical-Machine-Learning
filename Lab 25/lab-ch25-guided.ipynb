## Part 6: Reflection (10 min)

Answer each question in **3-5 sentences**. Connect your answers to concepts from Chapter 25.

### Q1: What worked well with AI-assisted development?

*Think about: speed of scaffolding, code quality, design suggestions, handling of repetitive tasks.*

**Your answer:** Claude Code scaffolded a complete Next.js portfolio site in under 4 minutes. If I make this portfolio manually which would spend me many hours. The generated code was clean, well-structured, and included components like dark/light mode toggle and responsive layout without any extra prompting. This reflects Chapter 25's idea that AI excels at repetitive, pattern-based tasks like boilerplate code generation, freeing the developer to focus on higher-level decisions like content and design.



### Q2: What did you have to fix manually?

*Think about: layout bugs, incorrect assumptions, styling mismatches, missing features, hallucinated APIs.*

**Your answer:** While the overall structure was satisfying, I had to manually correct the LinkedIn URL in the Hero and Contact sections, as Claude Code assumed a profile based on my name rather than knowing my actual account. I also had to provide real project descriptions, actual R² numbers from my lab outputs, and GitHub links — things Claude Code had no way of knowing. This shows that AI handles the how of building efficiently, but domain-specific personal details still require human verification and input.



### Q3: How does AI-assisted development compare to writing code entirely yourself?

*Think about: the HADI loop from Chapter 25 (Hypothesis, Action, Data, Insight). Did you follow this pattern? Where did you need domain knowledge that the AI lacked?*

**Your answer:** Following the HADI loop, I formed hypotheses about what prompts would produce good results, took action by pasting them into Claude Code, observed the output in the browser, and iterated with follow-up prompts to refine colors and content. Writing this entirely myself would have required knowing Next.js and other lots of skills that I don't know yet. The key difference is that AI-assisted development shifts the bottleneck from coding ability to clear communication and domain knowledge.



### Q4: Prompt engineering lessons

*What makes a good prompt for Claude Code? What makes a bad one? Give one specific example of a prompt you improved.*

**Your answer:** A good prompt is specific, detailed, and clearly describes the desired outcome, including exact content, colors, layout preferences, and links. A vague or messy prompt leaves Claude guessing, often producing generic results that need multiple rounds of correction. For example, my first project prompt used placeholder descriptions from the lab template; when I replaced them with my real R² numbers, actual GitHub links, and precise technology stacks, Claude Code generated much more accurate and personalized project cards. This reflects Chapter 25's idea that the quality of AI output is directly proportional to the clarity of human instruction.



### Q5: If you were hiring a junior data scientist, would their portfolio site influence your evaluation? Why or why not?

*Connect to the "signaling" concept from labor economics — what does a portfolio signal that a resume alone does not?*

**Your answer:** Yes, a portfolio site significantly influences my evaluation because it signals initiative, technical range, and the ability to ship a finished product — things a resume alone cannot demonstrate. In labor economics, signaling theory suggests that observable actions convey private information about a candidate's ability. A portfolio goes beyond credentials by showing real work: actual code, design judgment, and project outcomes, which are far harder to fake than a bullet point on a resume.
